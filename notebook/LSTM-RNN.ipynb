```python
## sampling and proofing the central limit theorm
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a non-normal dataset (random numbers from 1 to 100)
population = np.random.randint(1, 100, 10000)  # Large dataset of 10,000 values

# Step 2: Take many small samples and calculate their averages
sample_means = []
sample_size = 30  # Each sample will have 30 numbers
num_samples = 1000  # We'll take 1000 samples

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)  # Randomly pick 30 numbers
    print(sample)
    sample_means.append(np.mean(sample))  # Calculate and store the average

# Step 3: Plot the histogram of sample means
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem in Action")
plt.show()
```
